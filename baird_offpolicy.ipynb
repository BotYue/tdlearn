```python
import tdimport examplesimport numpy as npimport matplotlib.pyplot as pltfrom task import LinearValuePredictionTaskimport itertoolsnp.set_printoptions(precision=2, linewidth=120)def get_task(mu="target"):    n = 8    beh_pi = np.ones((n + 1, 2))    beh_pi[:, 0] = float(n) / (n + 1)    beh_pi[:, 1] = float(1) / (n + 1)    target_pi = np.zeros((n + 1, 2))    target_pi[:, 0] = 0    target_pi[:, 1] = 1    mdp = examples.BairdStarExample(n)    phi = mdp.phi    task = LinearValuePredictionTask(mdp, 0.99, mdp.phi,                                        np.asarray(n * [1.] + [10., 1.]),                                       policy=beh_pi, target_policy=target_pi)    task.beh_mu = np.array([1./(n+1)]*(n+1))    if mu is "target":        task.mu = np.array([0.] * n + [1.])    elif mu is "uniform":        task.mu = task.beh_mu.copy()    return task
```

## Parameter Search### Evaluation function

```python
from joblib import Parallel, delayedimport itertoolsfrom matplotlib.colors import LogNormdef plot_error_grid(val, alphas, mus, maxerr=5):    ferr = val.copy()    ferr[val > maxerr] = np.nan    plt.figure(figsize=(12,10))    plt.imshow(ferr, interpolation="nearest", cmap="hot", norm=LogNorm(vmin=np.nanmin(ferr), vmax=np.nanmax(ferr)))    plt.yticks(range(len(alphas)), alphas)    plt.xticks(range(len(mus)), mus, rotation=45, ha="right")    plt.xlabel(r"$\mu$")    plt.ylabel(r"$\alpha$")    plt.colorbar()
```
