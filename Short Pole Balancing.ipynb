```python
from joblib import Parallel, delayedimport itertoolsfrom matplotlib.colors import LogNormimport copydef plot_error_grid(val, alphas, mus, maxerr=5):    ferr = val.copy()    ferr[val > maxerr] = np.nan    plt.figure(figsize=(12,10))    plt.imshow(ferr, interpolation="nearest", cmap="hot", norm=LogNorm(vmin=np.nanmin(ferr), vmax=np.nanmax(ferr)))    plt.yticks(range(len(alphas)), alphas)    plt.xticks(range(len(mus)), mus, rotation=45, ha="right")    plt.xlabel(r"$\mu$")    plt.ylabel(r"$\alpha$")    plt.colorbar()
```

```python
def run(alpha, mu, method_class=td.GTD2):    t1 = copy.deepcopy(task)    np.seterr(all="ignore")    m = method_class(alpha=alpha, beta=mu*alpha, phi=t1.phi, gamma=gamma)    mean, std, raw = t1.avg_error_traces([m], n_indep=1, n_samples=50000, criterion="RMSPBE", verbose=False, error_every=500)    val = np.mean(mean[0,-70:])    #print "Evaluated alpha={}, mu={} to {}".format(alpha, mu, val)    return valgtd2alphas = np.linspace(0.1,0.8, 5) gtd2mus = [0.5, 0.1, 0.05, 0.01]k = (delayed(run)(*x) for x in itertools.product(gtd2alphas, gtd2mus))res = Parallel(n_jobs=4, verbose=11)(k) gtd2err = np.array(res).reshape((len(gtd2alphas), len(gtd2mus)))
```
